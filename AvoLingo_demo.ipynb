Import modules into model!

```python
import gradio as gr
from transformers import pipeline
```

Create a pipeline using Transformers. (English to French in this case)
[Documentation](https://huggingface.co/docs/transformers/main_classes/pipelines#transformers.TranslationPipeline)

```python
translation_pipeline = pipeline('translation_en_to_fr')
```

Test to see the translation

```python
translation_pipeline ('This is going to be a translation app!')
```

Make a wrapper function that takes in text (in this case English) and outputs the translated text (French!)

```python
def translate(raw_text):
    translated_text = translation_pipeline(raw_text)
    return translated_text[0]['translation_text']
```

Next wrap it all up in a [Gradio](https://www.gradio.app/) interface.

```python
interface = gr.Interface(fn=translate,
inputs=gr.components.Textbox(lines=2, placeholder='Text to translate'),
outputs='text') #specifies the output type to be text
```

```python
interface.launch()
```
